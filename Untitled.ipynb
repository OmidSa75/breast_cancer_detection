```python
import torch
import matplotlib.pyplot as plt
```

```python
mean = torch.tensor([0.])
std = torch.tensor([1.])
q = torch.distributions.Normal(mean, std)
```

```python
sample_data = q.rsample((100000,)).numpy()
```

```python
plt.figure(figsize=(15, 7))
plt.hist(sample_data, bins=100,rwidth=0.9)
plt.show()
```

```python
q2 = torch.distributions.Exponential(std)
```

```python
sample_data_2 = q2.rsample((100000,)).numpy()
```

```python
plt.figure(figsize=(15, 7))
plt.hist(sample_data_2, bins=100, rwidth=0.95)
plt.show()
```

```python
mean = torch.tensor([0.0, 10.0])
std = torch.tensor([1.0, 5.0])
q3 = torch.distributions.Normal(mean, std)
```
